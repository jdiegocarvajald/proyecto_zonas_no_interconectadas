# División de tablas para cargar en SQL

```python
# Se hace esta función porque siempre se pasa a csv de la misma forma
def df_to_csv(df, nombre):
    df.to_csv(f'{nombre}.csv', index=False, encoding='utf-8-sig')

# Al hacer Merges los Id's quedan tipo float64, esta función cambia los float64 de la tabla a int64
def float_to_int(df):
    for col in df.select_dtypes(include='float64').columns:
        df[col] = df[col].astype('Int64')
    return(df)
    
# Función para poner ID's serial como en SQL
def id_serial(df, col):
    df[col] = range(1, len(df) + 1) 
```

```python
departamentos = divapola[['Código Departamento', 'Nombre Departamento']].drop_duplicates().dropna()
df_to_csv(departamentos, 'departamentos')
departamentos.head()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1,1,1)
potencia.plot(ax=ax, legend=False)
plt.ylabel('POTENCIA MÁXIMA')
plt.savefig('mean_power.jpg')
plt.show()
```

```python
# Plot ACF and PACF
potencia_monthly=potencia.resample('ME').mean()
potencia_monthly.plot()
plt.show()
```

```python
first_order_diff=potencia_monthly.diff().dropna()
first_order_diff.info()
```

```python
first_order_diff.plot()
plt.show()
```

```python
plot_acf(first_order_diff)
plot_pacf(first_order_diff)
plt.show()
```

## ARMA

```python
#Optimize ARMA parameters
aicVal=[]
for ari in range(0,12):
    for maj in range(0,12):
        try:
            arima_obj = ARIMA(first_order_diff,order=(ari,0,maj))
            arima_obj_fit=arima_obj.fit()
            aicVal.append([ari, maj, arima_obj_fit.aic])
        except ValueError:
            pass
```

```python
aicVal.sort(key=lambda x:x[2])  # Ordena los valores de la tercera posición
aicVal
```

Se observa que AIRMA(0,1) es el más óptimo

```python
# Building optimized model using minimum AIC
arma_obj_fin = ARIMA(first_order_diff, order=(0, 1, 1)).fit()
```

```python
first_order_diff=pd.DataFrame(first_order_diff)
first_order_diff['ARMA(0,1)']=arma_obj_fin.predict()
```

```python
first_order_diff=first_order_diff.dropna()
first_order_diff['meantemp']=pd.to_numeric(first_order_diff['meantemp'])#.astype(float)
```

```python
first_order_diff=first_order_diff.dropna()
first_order_diff['POTENCIA MÁXIMA']=pd.to_numeric(first_order_diff['POTENCIA MÁXIMA'])#.astype(float)
```

```python
# Plot the curves
f, axarr = plt.subplots(1, sharex=True)
f.set_size_inches(12, 8)
first_order_diff['POTENCIA MÁXIMA'].iloc[1:].plot(color='b', label='Serie',linestyle = '-', ax=axarr)
first_order_diff['ARMA(0,1)'].iloc[1:].plot(color='r', label='Modelo ARMA',linestyle = '--', ax=axarr)
axarr.set_title('ARMA(2,3)')
axarr.legend()
plt.xlabel('Fecha de potencia máxima')
plt.ylabel('Potencia mácima') 
```
